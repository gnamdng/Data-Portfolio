#### Top 10 GDP Countries in 2023

```python
top10 = df.loc[df.index, '2023'].nlargest(10)
top10
```

```python
index_mapping = {
    186: "United States", 36: "China", 85: "Japan", 64: "Germany", 
    77: "India", 185: "United Kingdom", 60: "France",
    32: "Canada", 143: "Russian Federation", 24: "Brazil"     
}
top10.index = top10.index.to_series().replace(index_mapping)

top10 = top10.reindex(index_mapping.values())
top10.sort_values(ascending=False, inplace=True)
top10
```

```python
plt.figure(figsize=(12,5))
plt.bar(top10.index, top10)
plt.xlabel('Countries')
plt.ylabel('GDP')
plt.title('TOP 10 Countries GDP in 2023')
plt.show()
```

```python
df = df.rename(columns={"GDP, current prices (Billions of U.S. dollars)": "Country"})
```

```python
df.head(2)
```

```python
df2 = df.melt(id_vars=['Country'], var_name = "Year", value_name = "GDP")
print(df2)
```

Deleting useless last row

```python
index_to_drop = df2.loc[df2['Country'].str.contains('©IMF', na=False)].index

df2 = df2.drop(index=index_to_drop)
```

```python
df2.tail(2)
```

Last row was successfully deleted

Now changing Country Name for China so it will be easier and making a quick graph comparison between US, China and France.

```python
df2.loc[df2['Country'] == "China, People's Republic of", 'Country'] = "China"
```

```python
usachinafrance = df2[df2['Country'].isin(["China", "United States", "France"])]
plt.figure(figsize=(25,12))
sns.lineplot(data = usachinafrance, x='Year', y='GDP', hue="Country")
plt.grid()
plt.show()
```

#### Top 10 GDP Lineplot

Graph of the GDP evolution of the top 10 GDP countries. Much more pleasant to the eye

```python
top_countries = ['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France', 'Canada', 'Russian Federation', 'Brazil']
df_top = df2[df2['Country'].isin(top_countries)]

plt.figure(figsize=(23, 12))
sns.lineplot(data=df_top, x="Year", y="GDP", hue="Country")
plt.xlabel("Year")
plt.ylabel("GDP in billions")
plt.xticks(rotation=45) 
plt.grid()
plt.show()
```

Overall all top 10 GDP countries are stable and under 5000 billions, except for Japan going slightly up and down are the value.
As for the US and China, their GDP is skyrocketing, tripling over the course of 15 years, China is starting to close the gap with US.
Russia which is named Russian Federation started at 1989 as it was the end of URSS.

```python
df = df.sort_values(by=['Country', 'Year'])

df['GDP_growth_rate'] = df.groupby('Country')['GDP'].pct_change() * 100
df['GDP_growth_rate'] = df['GDP_growth_rate'].fillna(0)
df_pivot = df.pivot(index='Country', columns='Year', values='GDP_growth_rate')
df_pivot
```

Let's compare growth rate of US and China.

```python
uschina = df_pivot.loc[['China', 'United States']]
uschina
```

```python
plt.figure(figsize=(25,10))
sns.lineplot(data=uschina.T[['United States',"China"]],marker = 'o')
plt.ylabel('Growth Rate (%)')
plt.title('Growth Rate Comparison between US and China')
plt.show()
```
